```python
mydata = pd.merge(new_sales, new_inventory, left_on =['date', 'store_id'], 
                  right_on=['year_month', 'Store_ID'], how='left')
```

```python
mydata = mydata[['date','store_id', 'warehouse','transactions','revenue', 'net_revenue',
                 'qty', 'unit_cost', 'unit_price', 'total_cost', 'total_retail']]
mydata.tail()
```

```python
mydata.set_index('date', inplace=True)
mydata.sort_index(inplace=True)
```

```python
filter_mydata = mydata.loc['2021':'2024']
filter_mydata.head()
```

```python
monthly_summary = filter_mydata.groupby(['date', 'store_id']).agg({
    'transactions': 'nunique',
    'revenue': 'sum',
    'qty': 'sum'
})

monthly_summary
```

```python
## Stores as Columns
monthly_table = monthly_summary.pivot_table(index='date', columns='store_id', values=['transactions','revenue','qty'])
monthly_table.tail()
```
